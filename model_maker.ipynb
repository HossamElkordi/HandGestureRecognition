# Create Dataset

```python
datagen_train = ImageDataGenerator(rescale=1.0/255.0, dtype='float32', validation_split=0.2)
img_size = [96, 96]
```

```python
train_dataset = datagen_train.flow_from_directory(
    data_dir,
    target_size=img_size,
    color_mode='grayscale',
    batch_size=32,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

validation_dataset = datagen_train.flow_from_directory(
    data_dir,
    target_size=img_size,
    color_mode='grayscale',
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

# Build Model

```python
def build_model():
    model = models.Sequential([
        layers.Input(shape=(*img_size, 1)),
        layers.Conv2D(32, kernel_size=(3, 3), activation='relu', data_format='channels_last'),
        layers.DepthwiseConv2D(kernel_size=(3, 3), activation='relu', data_format='channels_last'),
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dense(4, activation='softmax')
    ])
    return model

model = build_model()
model.summary()
```

```python
def train_model(model):
  batch_size = 128
  epochs = 20
  lr = 0.003
  scheduler = tf.keras.optimizers.schedules.CosineDecay(initial_learning_rate=lr, alpha=0.0, decay_steps=epochs * len(train_dataset))
  opt = tf.keras.optimizers.Adam(learning_rate=scheduler)
  model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
  history = model.fit(train_dataset, validation_data=validation_dataset, epochs=epochs, verbose=1)
  return model, history
```

```python
def plot_history(history):
  plt.figure(figsize=(10, 4))
  plt.subplot(1, 2, 1)
  plt.plot(history.history['loss'], label='Training Loss')
  plt.plot(history.history['val_loss'], label='Validation Loss')
  plt.legend()

  plt.subplot(1, 2, 2)
  plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
  plt.legend()
  plt.show()
```

```python
trained_model, history = train_model(model)
```

```python
model.evaluate(validation_dataset)
```

```python
plot_history(history)
```

# Save Model

```python
quant_train_dataset = datagen_train.flow_from_directory(
    data_dir,
    target_size=img_size,
    color_mode='grayscale',
    batch_size=1,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

eval_dataset = datagen_train.flow_from_directory(
    data_dir,
    target_size=img_size,
    color_mode='grayscale',
    batch_size=1,
    class_mode='categorical',
    subset='validation',
    shuffle=True
)
```

```python
def representative_dataset_gen():
  for i in range(len(quant_train_dataset)):
      image = quant_train_dataset[i][0]
      yield [image]

def convert_to_tflite(model, quantize=False):
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    if quantize:
        converter.optimizations = [tf.lite.Optimize.DEFAULT]
        converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
        converter.representative_dataset = tf.lite.RepresentativeDataset(representative_dataset_gen)
        converter.inference_input_type = tf.uint8
        converter.inference_output_type = tf.uint8
    return converter.convert()

def save_tflite_model(tflite_model, filename):
    with open(filename, 'wb') as f:
        f.write(tflite_model)
```

```python
tflite_model = convert_to_tflite(trained_model, quantize=False)
save_tflite_model(tflite_model, '/home/h.elkordi/projects/EdgeML/Project/models/model.tflite')

quantized_tflite_model = convert_to_tflite(trained_model, quantize=True)
save_tflite_model(quantized_tflite_model, '/home/h.elkordi/projects/EdgeML/Project/models/quantized_model.tflite')
```

# Profile Model

```python
def evaluate_model(tflite_model, eval_data):
    interpreter = tf.lite.Interpreter(model_content=tflite_model)
    interpreter.allocate_tensors()

    input_details = interpreter.get_input_details()[0]
    output_details = interpreter.get_output_details()[0]

    input_scale, input_zero_point = input_details["quantization"]
    output_scale, output_zero_point = output_details["quantization"]

    correct_predictions = 0
    for i in range(len(eval_data)):
        input_data = eval_dataset[i][0]
        if (input_scale, input_zero_point) != (0.0, 0):
          input_data = input_data / input_scale + input_zero_point
          input_data = input_data.astype(input_details["dtype"])
        interpreter.set_tensor(input_details["index"], input_data)
        interpreter.invoke()
        predictions = interpreter.get_tensor(output_details["index"])

        if (output_scale, output_zero_point) != (0.0, 0):
          predictions = predictions.astype(np.float32)
          predictions = (predictions - output_zero_point) * output_scale

        if np.argmax(predictions) == np.argmax(eval_dataset[i][1]):
            correct_predictions += 1

    accuracy = correct_predictions / len(eval_data)
    return accuracy
```

```python
model_acc = evaluate_model(tflite_model, eval_dataset)
quantized_model_acc = evaluate_model(quantized_tflite_model, eval_dataset)
```

```python
print('Model Accuracy:', model_acc)
print('Quantized Model Accuracy:', quantized_model_acc)
```

```python
def profile_model(model, quant_model, eval_data):
    start_time = time.time()
    evaluate_model(model, eval_data)
    model_time = time.time() - start_time

    start_time = time.time()
    evaluate_model(quant_model, eval_data)
    quant_model_time = time.time() - start_time

    return 1000 * model_time / len(eval_data), 1000 * quant_model_time / len(eval_data)
```

```python
model_time, quantized_model_time = profile_model(tflite_model, quantized_tflite_model, eval_dataset)
```

```python
print('Model Inference Time:', model_time, 'ms')
print('Quantized Inference Time:', quantized_model_time, 'ms')
```
